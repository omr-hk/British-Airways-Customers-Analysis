```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('../dataset/customer_booking.csv',encoding="ISO-8859-1")
data.head()
```

```python
data.corr(numeric_only=True)
```

```python
sns.pairplot(data)
```

```python
data.info()
```

```python
data['booking_complete'].value_counts()
```

```python
cat=[col for col in data.columns if data[col].dtype==object]
metric=[col for col in data.columns if col not in cat]
print(cat)
print(metric)
```

```python
df=pd.concat([data[metric],pd.get_dummies(data[cat],drop_first=True)],axis=1)
```

```python
df.info()
```

```python
cols=[col for col in df.columns if col!='booking_complete']
```

```python
x=df[cols]
y=df['booking_complete']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=7)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier=RandomForestClassifier(random_state=7)
classifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_test_pred=classifier.predict(x_test)
y_train_pred=classifier.predict(x_train)
print('test: ',accuracy_score(y_test,y_test_pred))
print('train: ',accuracy_score(y_train,y_train_pred))
```

```python
from skopt import BayesSearchCV
```

```python
classifier1=RandomForestClassifier(random_state=7)
params={
    'n_estimators':(100,500),
    'max_depth':(3,10),
    'max_features':(0.1,1),
}
search=BayesSearchCV(classifier1,params,n_jobs=-1)
```

```python
search.fit(x_train,y_train)
```

```python
search.score(x_test,y_test)
```

```python
search.score(x_train,y_train)
```

```python
search.best_params_
```

```python
classifier_main=RandomForestClassifier(max_features=1.0,max_depth=10,n_estimators=453,random_state=7)
classifier_main.fit(x_train,y_train)
```

```python
y_test_pred=classifier_main.predict(x_test)
y_train_pred=classifier_main.predict(x_train)
print('test result: ',accuracy_score(y_test,y_test_pred))
print('train result: ',accuracy_score(y_train,y_train_pred))
```

```python
features=classifier_main.feature_importances_
indices=np.argsort(features)[::-1]
columns=x_train.columns
for f in range(x_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30,
                            columns[indices[f]],
                            features[indices[f]]))
```

```python
fig,ax=plt.subplots(figsize=(25,10))
ax.set_title("Feature Importance")
ax.bar(range(100), features[indices[:100]], align='center')
ax.set_xticks(range(100), x_train.columns[indices[:100]], rotation=90)
plt.tight_layout()
plt.savefig('../plots/feature-importance.png')
```
